```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xy = np.loadtxt("Helix.csv",delimiter = ",")
print(xy.shape)
xy
```

```python
def tranf(pt,A):
    newp = []
    for xy in pt:
        newp.append(A@xy)
    return np.array(newp)
```

```python
plt.plot(xy[0],xy[1],'bo',markersize = 15)
plt.show()
```

# New(ex)

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
#ax = fig.gca(projection = '3d')
ax = fig.add_subplot(projection='3d')
plt.plot(xy[:,0], xy[:,1], xy[:,2],'.g')
 
th = 11*np.pi/6
A = np.array([[np.cos(th), -np.sin(th), 0],
             [np.sin(th), np.cos(th), 0],
             [0, 0, 1]])
newxy = tranf(xy, A)
plt.plot(newxy[:,0], newxy[:,1], newxy[:,2],'m')
```
