```python
xy = np.array([[1],[1]])
A = np.array([[1,0],[0,1]])
print(xy)
print(A)
newxy = A@xy
plt.plot(xy[0],xy[1],'bo',markersize = 15)
plt.plot(newxy[0],newxy[1],'r>',markersize = 15)
plt.show()
```

```python
xy = np.loadtxt("data.csv",delimiter = ",")
xy.shape
```

```python
plt.plot(xy[:,0],xy[:,1])
```

```python
def tran(pt,A):
    newp = []
    for xy in pt :
        newp.append(A@xy)        
    return np.array(newp)
```

```python
A = np.array([[1,0],[0,-1]])
newxy = tran(xy,A)
plt.plot(xy[:,0],xy[:,1])
plt.plot(newxy[:,0],newxy[:,1],'r')
```

```python
A = np.array([[-1,0],[0,-1]])
newxy = tran(xy,A)
plt.plot(xy[:,0],xy[:,1])
plt.plot(newxy[:,0],newxy[:,1],'r')
```

```python
th = np.pi/2
A = np.array([[np.cos(th),-np.sin(th)],[np.sin(th),np.cos(th)]]) # [0,-1],[1,0]
newxy = tran(xy,A)
plt.plot(xy[:,0],xy[:,1])
plt.plot(newxy[:,0],newxy[:,1],'r')
```

```python
th = np.pi/6
A = np.array([[np.cos(th),-np.sin(th)],[np.sin(th),np.cos(th)]]) # [0,-1],[1,0]
newxy = tran(xy,A)
plt.plot(xy[:,0],xy[:,1])
plt.plot(newxy[:,0],newxy[:,1],'r')
```

```python
plt.plot(xy[:,0],xy[:,1])
for k in range(12):
    B = np.array([[np.cos(np.pi*k/6),-np.sin(np.pi*k/6)],[np.sin(np.pi*k/6),np.cos(np.pi*k/6)]])
    newxy = tran(xy,B)
    plt.plot(newxy[:,0],newxy[:,1])
plt.show
```

```python
plt.plot(xy[:,0],xy[:,1])
for k in range(12):
    B = np.array([[np.cos(np.pi*k/6),-np.sin(np.pi*k/6)],[np.sin(np.pi*k/6),np.cos(np.pi*k/6)]])
    newxy = tran(xy,B)
    if k%2==0 :
        plt.plot(newxy[:,0],newxy[:,1],'g')
    else :
        plt.plot(newxy[:,0],newxy[:,1],'m')
plt.show
```
